```python
# (i)Is there any relationship in between labour force status with marital status
```

```python
# we need to verify reltionship between two categorical variables so we will use Chi square test
# H0: the categorical variables are not related
# H1: the categorical variables are related
```

```python
C = pd.crosstab(sample_survey.wrkstat ,sample_survey.marital , margins = True)
C
```

```python
# (ii)Is there any relationship in between educational qualification with marital status
```

```python
# we need to verify reltionship between two categorical variables so we will use Chi square test
# H0: the categorical variables are not related
# H1: the categorical variables are related
```

```python
C = pd.crosstab(sample_survey.degree ,sample_survey.marital , margins = True)
C
```

```python
stats.chi2_contingency(observed = C)
```

```python
# Since p values are less than 5%  they we will reject the null Hypothesis Hence, they are related
```
